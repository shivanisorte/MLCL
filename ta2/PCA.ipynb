#### PCA
- It stands dor Principal Component Analysis.
- It is a Unsupervised Machine Learning Techniques and widely used in Dimensionality Reduction.

#### What is Dimensionality Reduction ?
- It is a mathematical process which reduce the Dimension of the data (Dimension is nothing but variable/feature if dataset).
- For example, if we have 100 dimension in our dataset contributing to Target feature, using all 100 features for prediction of Target feature nothing make sense as some of feaetures are not contributing to target features and some feature are corelated with each other which gives same information for prediction, hence using 10 Principal component for prediction is more effecting than using 100 features.

#### PCA Implementation Steps
1. PCA uses mathematical model which works on Mean and variance to reduce n feature into 1 PC.
2. PCA involves : standardization of data with mean 0 and variance 1 which bring all data in range of -1 to 1 with normal distribution.
3. then it computes the coveriance matrix of features.
4. from covariance matrix it obtains the Eigen vectors and Eigan values.
5. it sort the eigan value in descending order and choose the top m values using distance formula.
6. those top m values are called as Principal component.



#### Applications of PCA in Machine Learning
- PCA is used to visualize multidimensional data.
- It is used to reduce the number of dimensions in healthcare data.
- PCA can help resize an image.
- It can be used in finance to analyze stock data and forecast returns.
- PCA helps to find patterns in the high-dimensional datasets.


#### Process of PCA



1. Normalize the data
- Standardize the data before performing PCA. This will ensure that each feature has a mean = 0 and variance = 1.



2. Build the covariance matrix
- Construct a square matrix to express the correlation between two or more features in a multidimensional dataset.



3. Find the Eigenvectors and Eigenvalues
- Calculate the eigenvectors/unit vectors and eigenvalues. Eigenvalues are scalars by which we multiply the eigenvector of the covariance matrix.



4. Sort the eigenvectors in highest to lowest order and select the number of principal components.

```python
def draw_roc( actual, probs ):
    fpr, tpr, thresholds = metrics.roc_curve( actual, probs,
                                              drop_intermediate = False )
    auc_score = metrics.roc_auc_score( actual, probs )
    plt.figure(figsize=(6, 6))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()

    return fpr, tpr, thresholds
```

### PCA on the data

#### Note - 
- While computng the principal components, we must not include the entire dataset. Model building is all about doing well on the data we haven't seen yet!
- So we'll calculate the PCs using the train data, and apply them later on the test data

```python
X_train.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(random_state=42)
```

```python
pca.fit(X_train)
```

#### Components from the PCA

Looking at the explained variance ratio for each component

```python
pca.components_
```

```python
pca.components_.shape
```

```python
pca.explained_variance_ratio_
```

Making a scree plot for the explained variance

```python
var_cumu = np.cumsum(pca.explained_variance_ratio_)
```

```python
fig = plt.figure(figsize=[12,8])
plt.vlines(x=15, ymax=1, ymin=0, colors="r", linestyles="--")
plt.hlines(y=0.95, xmax=30, xmin=0, colors="g", linestyles="--")
plt.plot(var_cumu)
plt.ylabel("Cumulative variance explained")
plt.show()
```

Perform PCA with 16 components

```python
from sklearn.decomposition import IncrementalPCA
```

```python
pca_final = IncrementalPCA(n_components=16)
```

```python
df_train_pca =  pca_final.fit_transform(X_train)
```

```python
df_train_pca.shape
```

Plotting the heatmap of the corr matrix

```python
corrmat   = np.corrcoef(df_train_pca.transpose())
```

Applying the transformation on the test set

```python
corrmat.shape
```

```python
plt.figure(figsize=[16,16])
sns.heatmap(corrmat, annot=True)
```

```python
df_test_pca = pca_final.transform(X_test)
```

```python
df_test_pca.shape
```

### Applying logistic regression on the data on our Principal components

```python
from sklearn.linear_model import LogisticRegression
```

```python
learner_pca = LogisticRegression()
```

```python
model_pca = learner_pca.fit(df_train_pca, y_train)
```

Making predictions on the test set

```python
pred_probs_test = model_pca.predict_proba(df_test_pca)
```

```python
"{:2.2}".format(metrics.roc_auc_score(y_test, pred_probs_test[:,1]))
```
